```python
def add_zero(num):
    if (int(num) < 10):
        return "0" + str(num)
    return str(num)
```

```python
def get_date(cal):
    month = cal.find_element(By.XPATH,".//span[@class='month']").text
    day = cal.find_element(By.XPATH,".//span[@class='day']").text
    num_month = add_zero(uppercase_months.index(month))
    num_day = add_zero(day)
    if(12 >= int(num_month) >= 10):
        return num_day+"-"+num_month+"-2024"
    else:
        return num_day+"-"+num_month+"-2025"

def get_week_day(cal):
    return cal.find_element(By.XPATH,".//span[@class='dayname']").text
```

```python
def check_am_pm(string):
    if string.find("am") >= 0 or string.find("12") >= 0:
        return string[0:string.find("m")-1]
    else:
        hour = 12 + int(string[0:string.find(":")])
        return str(hour) + string[string.find(":"):-2]
```

```python
def get_start_time(event_time):
    time_list = event_time.split(" ")
    return check_am_pm(time_list[0])


def get_end_time(event_time):
    time_list = event_time.split(" ")
    return check_am_pm(time_list[2])
```
